```r
boot.lm=function(dtx,B)
{ 
    dt=data.frame(dtx)
    dt1=dt[,1]
    dt2=dt[,2]
    theta.hat=lm(dt1~dt2,data=dtx)$coef[2]
    n=nrow(dtx)
    theta.b=numeric(B)
    set.seed(100)
    for (b in 1:B)
    { 
        k= sample(1:n,n,replace=T)
        dt1.1=dt1[k]
        dt2.1=dt2[k]
        theta.b[b]=lm(dt1.1~dt2.1)$coef[2]
    }
    bias=mean(theta.b-theta.hat)
    se=sd(theta.b)
    list(sample.boot=theta.b[1:20],est=theta.hat,bias=bias,se=se)
}
boot.lm(women,1000)
```

```r
jack.fn=function(dtx) 
{ 
    dt=data.frame(dtx)
    dt1=dt[,1]
    dt2=dt[,2]
    theta.hat=cor(dt1,dt2)
    n=nrow(dtx)
    theta.jack=numeric(n)

    for (i in 1:n)
    { 
        theta.jack[i]=cor(dt1[-i],dt2[-i])
    }

    bias=(n-1)*(mean(theta.jack)-theta.hat)
    se=sqrt((n-1)*mean((theta.jack-mean(theta.jack))^2))
    list(est=theta.hat,bias=bias,se=se)
}

jack.fn(women)
```
