```r
n=2000
x1=rgamma(n,2,2)
x2=rgamma(n,2,4)
s=x1+x2
par(mfrow=c(2,1))
hist(s,prob=T)
u=runif(n)
u[1:20]
k=as.integer(u>=0.5)
x=k*x1+(1-k)*x2
hist(x,prob=T)
```

```r
n=5000
p=c(0.1,0.2,0.2,0.3,0.2)
lambda=c(1,1.5,2,2.5,3)
k=sample(1:5,size=n,replace=T,prob=p)
k[1:20]
rate=lambda[k]
rate[1:20]
x=rgamma(n,shape=3,rate=rate)
plot(density(x),lwd=4,col="maroon",ylim=c(0,0.8))
for (i in 1:5)
    lines(density(rgamma(n,3,lambda[i])))
```

```r
fu.dens=function(x,lambda,theta)
{
    sum(dgamma(x,3,lambda)*theta)
}
x=seq(0,8,length=200)
dim(x)=length(x)
y=apply(x,1,fu.dens,lambda=lambda,theta=p)
plot(x,y,type="l",lwd=4,col="orange", ylim=c(0,0.9))
for (i in 1:5)
{
    y=apply(x,1,dgamma,shape=3,rate=lambda[i])
    lines(x,y)
        
}
```
